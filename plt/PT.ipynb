```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = 1j * np.linspace(0.001,1000,1000000)
tau = 0.4
G = 1/(1+tau*s)
```

```python
fig = plt.figure()
fig.set_figwidth(6)
fig.set_figheight(4)

plt.subplot(211)
plt.title("Frequenzgang")
plt.axvline(tau*2*np.pi, linestyle = ":", color = "grey")
plt.axhline(-3, linestyle = ":", color = "grey")
plt.plot(np.imag(s), 20*np.log10(np.abs(G)), color = "red")
plt.xscale("log")
plt.ylim(-40,5)
plt.grid(linestyle = ":",which="both",alpha = 0.3)

plt.subplot(212)
plt.title("Phasengang")
plt.axvline(tau*2*np.pi, linestyle = ":", color = "grey")
plt.axhline(-np.pi/4, linestyle = ":", color = "grey")
plt.plot(np.imag(s), np.angle(G), color = "orange")
plt.xscale("log")
plt.grid(linestyle = ":",which="both",alpha = 0.3)

plt.tight_layout()
plt.savefig("PT1-Freq.pdf")
```

```python
omega1 = 10
d1 = 0.1
H1 = omega1**2 / ( s**2 + 2*d1*omega1*s+omega1**2 )
omega2 = 10
d2 = 0.6
H2 = omega2**2 / ( s**2 + 2*d2*omega2*s+omega2**2 )

fig = plt.figure()
fig.set_figwidth(6)
fig.set_figheight(4)

plt.subplot(211)
plt.title("Frequenzgang")
plt.axvline(tau*2*np.pi, linestyle = ":", color = "grey")
plt.axhline(-3, linestyle = ":", color = "grey")
plt.plot(np.imag(s), 20*np.log10(np.abs(H1)), color = "red")
plt.plot(np.imag(s), 20*np.log10(np.abs(H2)), color = "orange")
plt.xscale("log")
plt.ylim(-40,15)
plt.grid(linestyle = ":",which="both",alpha = 0.3)

plt.subplot(212)
plt.title("Phasengang")
plt.axvline(tau*2*np.pi, linestyle = ":", color = "grey")
plt.axhline(-np.pi/4, linestyle = ":", color = "grey")
plt.plot(np.imag(s), np.angle(H1), color = "red", label = "d = 0.1")
plt.plot(np.imag(s), np.angle(H2), color = "orange", label = "d = 0.6")
plt.xscale("log")
plt.grid(linestyle = ":",which="both",alpha = 0.3)

plt.legend(title = "Dmpfungen")

plt.tight_layout()
plt.savefig("PT2-Freq.pdf")
```
